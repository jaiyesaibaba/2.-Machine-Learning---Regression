```python
# pandas is a library and it contains functions related to the database
import pandas as pd
```

```python
dataset=pd.read_csv("Salary_Data.csv")
#dataset is called by read_csv function under pandas and it is assigned to the dataset variable
```

```python
dataset
```

```python
dataset
#checking the data present inside it
```

```python
# assigning dependent and independent variable
# assigning a particular column to a variable for example assigning years of experience column to the independent variable
independent = dataset[["YearsExperience"]]
```

```python
independent
#checking the data saved in this particular variable
```

```python
dependent = dataset[["Salary"]]
#assigned the salary column from dataset variable to the dependent varaiable
```

```python
dependent
#checking the data present inside the dependent variable
```

```python
#splitting training set and test set
#using sklearn library which supports machine learning we ar splitting using the following procedure
from sklearn.model_selection import train_test_split
#loaded the function
#here file name is sklearn.model_selection and function or class name is train_test_split
#here sklearn is a folder and model_selection is a function
#it saves the memory
#train_test_split acts as a parameterized function and we are going to pass the input and output
#train_test_split(independent,dependent,test_size=0.30,random_state=0)
#passed the parameters
#we are going to assign it to the variable so that we can use it for the further process
x_train,x_test,y_train,y_test =train_test_split(independent,dependent,test_size=0.30,random_state=0)
#we are going to get the output for all the four variable passed above 
```

```python
x_train,x_test,y_train,y_test
#x means input and y means output maximum percentage of data alloted to training set and minimum datas allowed for test set only when we split like this we can access the sklearn
```

```python
from sklearn.linear_model import LinearRegression
#above line is going to store in the memory and it got loaded it is going to exeute the next line
regressor=LinearRegression()
#above two lines are the procedure
#below is for substitution
regressor.fit(x_train,y_train)
#in the above line only using fit method we are just subtituting the values for the formula
#above line after executing that weight and bias got understood
#model creation done here after execution
#linearregression means weight and bias which means slope and intercept
```

```python
weight=regressor.coef_
weight
```

```python
#the above value is the prediction for the next step
bias = regressor.intercept_
bias
```

```python
#above value is the origin value that is 0 years of experience
#in regressor only all the values got stored likeweight, bias etc.,
y_pred=regressor.predict(x_test)
```

```python
#all the test values got stored inside y_pred variable
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```

```python
r_score
#r_score is nearly 1 means good model
```

```python
#so it is a good model according to the output
#upcoming step we aare storing the model
import pickle
filename="finalized_model_linear.sav"
```

```python
pickle.dump(regressor,open(filename,'wb'))
```

```python
#loading the model
loaded_model=pickle.load(open("finalized_model_linear.sav",'rb'))
result=loaded_model.predict([[15]])
result
```
